```python
dff.drop(['employment_industry', 'employment_occupation', 'health_insurance'], axis=1, inplace=True)
dft.drop(['employment_industry', 'employment_occupation', 'health_insurance'], axis=1, inplace=True)
```

```python
for category in dff.columns:
  dff[category].fillna(dff[category].mode()[0])
  dft[category].fillna(dft[category].mode()[0])
```

```python
for category in dff.columns:
  dff[category] = dff[category].astype("category").cat.codes
  dft[category] = dft[category].astype("category").cat.codes
```

```python
dff
```

```python
dff1 = dff.drop(['doctor_recc_seasonal', 'opinion_seas_vacc_effective', 'opinion_seas_risk', 'opinion_seas_sick_from_vacc'], axis=1)
dft1 = dft.drop(['doctor_recc_seasonal', 'opinion_seas_vacc_effective', 'opinion_seas_risk', 'opinion_seas_sick_from_vacc'], axis=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
m1 = RandomForestClassifier()
m1.fit(dff1, dfl['xyz_vaccine'])
```

```python
dff2 = dff.drop(['xyz_concern', 'xyz_knowledge', 'doctor_recc_xyz', 'opinion_xyz_vacc_effective', 'opinion_xyz_risk', 'opinion_xyz_sick_from_vacc'], axis=1)
dft2 = dft.drop(['xyz_concern', 'xyz_knowledge', 'doctor_recc_xyz', 'opinion_xyz_vacc_effective', 'opinion_xyz_risk', 'opinion_xyz_sick_from_vacc'], axis=1)
```

```python
m2 = RandomForestClassifier()
m2.fit(dff2, dfl['seasonal_vaccine'])
```

```python
p1 = m1.predict_proba(dft1)[:, 1]
p2 = m2.predict_proba(dft2)[:, 1]
```

```python
r = pd.read_csv('submission_format.csv', index_col='respondent_id')
r.drop(['h1n1_vaccine', 'seasonal_vaccine'], axis=1, inplace=True)
r.insert(0, 'h1n1_vaccine', p1, False)
r.insert(1, 'seasonal_vaccine', p2, False)
r
```

```python
r.to_csv('final_submission.csv')
```
